```python
import numpy as np
import pandas as pd
```

```python
#1.Query: to filter a data frame
shoesize_1 = np.random.randint(10,size=10)
dressize_2= np.random.randint(10,size=10)
years=np.arange(2010,2020)
groups=["A","B","A","C","A","B","B","C","C","A"]

df=pd.DataFrame({"group":groups,"year":years,"shoesize_1":shoesize_1,"dressize_2":dressize_2})

df
```

```python
#Query is used to query a function which only requires to write the condition for filtering
df.query("shoesize_1<dressize_2")
```

```python
#2. Insert: to add a new column to a dataframe. it is usually added at the end by default.
#new column
new_col=np.random.randn(10)

#insert the new column at position 2
df.insert(2, "new_col",new_col)

df
```

```python
#sample: allows you to select values randomly from a series or data frame
sample1 = df.sample(n=3)
sample1
```

```python
sample2 = df.sample(frac=0.5)
sample2
```

```python
#where: used to replace values in rows or column based on a condition
df["new_col"].where(df["new_col"]>0,0)
```

```python
df["new_col"].where(df["new_col"]>0,0)
np.where(df["new_col"]>0,df["new_col"],0)
```

```python
#Isin: to filter or select methods a lot when working with data frames. it is an advanced filetring
years = ["2011","2013","2019"]
df[df.year.isin(years)]
```

```python
#LOC:to select data by labels
df.loc[:2,["group","year"]]
```

```python
#ILOC:to select data by positions
df.iloc[:3,:2]
```

```python
#Pct_change: to calculate percentage  through the values in a series
df.shoesize_1.pct_change()
```

```python
#Rank:assigns rank to the values
df["rank_1"] = df["shoesize_1"].rank()
df
```
